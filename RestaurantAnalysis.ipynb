```python
def read_files(file_path , encodings=['utf-8', 'latin-1', 'ISO-8859-1']):
    for enc in encodings:
        try:
            return pd.read_csv(file_path , encoding = enc)
        except UnicodeDecodeError:
            continue
    raise ValueError(f"Unable to read {file_path} with any of the specified encodings.")
            
```

```python
menuItems_df = read_files(menuItems)
orderDetails_df = read_files(orderDetails)
res_db_df = read_files(res_db)
```
